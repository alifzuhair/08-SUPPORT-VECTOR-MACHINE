```python
# 1. DOWNLOAD & LOAD DATASET
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

url_dataset = 'https://raw.githubusercontent.com/alifzuhair/08-SUPPORT-VECTOR-MACHINE/main/spam_dataset_small.csv'

print("Sedang mengunduh dataset dari GitHub...")
df = pd.read_csv(url_dataset)

print("--- 5 Data Teratas ---")
print(df.head())
```

### 1. Load Dataset

Membuka dataset yang isinya daftar teks email panjang beserta label apakah itu spam atau tidak.

```python
# 2. PREPROCESSING DATA
print("\n--- Memeriksa Missing Value ---")
print(df.isnull().sum())

df = df.dropna()

X_raw = df['text']
y = df['label'] # 1 = Spam, 0 = Bukan Spam (Ham)

vectorizer = TfidfVectorizer(max_features=5000)
X_processed = vectorizer.fit_transform(X_raw)
```

### 2. Preprocessing Data

Mengecek dan membuang jika ada baris email yang kosong (missing value).

Menggunakan metode bernama TF-IDF Vectorizer. Metode ini melakukan dua instruksi tugas sekaligus: mengubah kata-kata teks menjadi bentuk angka (Encoding) dan langsung menormalkan bobot rentang angkanya (Scaling) berdasarkan seberapa penting kata tersebut di dalam email.

```python
# 3. SPLIT DATA

X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)
print(f"\nJumlah data latih: {X_train.shape[0]} | Jumlah data uji: {X_test.shape[0]}")
```

### 3. Split Data

Membagi tumpukan email tersebut menjadi dua keranjang: 80% untuk bahan belajar si SVM (train data), dan 20% sisanya disembunyikan untuk bahan ujian (test data).

```python
# 4. MODELING SVM

model = SVC(kernel='linear', random_state=42)

print("\nSedang melatih model SVM (Tunggu sebentar, data cukup besar)...")
model.fit(X_train, y_train)
print("Selesai melatih model!")
```

### 4. Modeling SVM

Mengubah pilihan dari Kernel RBF menjadi Kernel Linear. Setelah teks diubah menjadi angka, jumlah fiturnya akan membengkak menjadi ribuan (sebanyak jumlah kosakata kata). SVM dengan Kernel Linear adalah yang paling optimal, cepat, dan anti-overfitting untuk menangani data teks berdimensi tinggi.

```python
# 5. EVALUASI MODEL

y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

acc_train = accuracy_score(y_train, y_pred_train)
acc_test = accuracy_score(y_test, y_pred_test)

print("\n================ EVALUASI MODEL ================")
print(f"Accuracy  : {accuracy_score(y_test, y_pred_test):.4f}")
print(f"Precision : {precision_score(y_test, y_pred_test, average='binary'):.4f}")
print(f"Recall    : {recall_score(y_test, y_pred_test, average='binary'):.4f}")
print(f"F1 Score  : {f1_score(y_test, y_pred_test, average='binary'):.4f}")
print("================================================")

print("\n--- Analisis Singkat ---")
print(f"Akurasi Data Latih (Train): {acc_train:.4f}")
print(f"Akurasi Data Uji (Test)   : {acc_test:.4f}")
if acc_train - acc_test > 0.05:
    print("Kesimpulan: Model ada kecenderungan OVERFITTING (Bagus di data latih, tapi menurun di data uji).")
elif acc_train < 0.60 and acc_test < 0.60:
    print("Kesimpulan: Model UNDERFITTING (Gagal mengenali pola spam).")
else:
    print("Kesimpulan: Model GOOD FIT (Kinerjanya sangat seimbang dan memuaskan pada data latih maupun uji).")
```

### 5. Evaluasi Model

Menghitung nilai ujian si SVM (Akurasi, Presisi, dll) untuk melihat kemampuannya mendeteksi spam, sekaligus membandingkan nilai saat latihan vs nilai saat ujian untuk mengetahui apakah dia overfitting atau sudah pas (good fit).

```python
# 6. VISUALISASI
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
label_names = y.map({0: 'Bukan Spam (0)', 1: 'Spam (1)'})
sns.countplot(x=label_names, palette='Set2')
plt.title('Distribusi Target (Spam vs Bukan Spam)')
plt.xlabel('Kategori Email')
plt.ylabel('Jumlah Email')

plt.subplot(1, 2, 2)
cm = confusion_matrix(y_test, y_pred_test)
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', xticklabels=['Bukan Spam', 'Spam'], yticklabels=['Bukan Spam', 'Spam'])
plt.title('Confusion Matrix')
plt.xlabel('Tebakan Model (Prediksi)')
plt.ylabel('Kenyataan (Aktual)')

plt.tight_layout()
plt.show()
```

### 6. Visualisasi

Grafik batang untuk melihat perbandingan jumlah email spam dan email asli di dataset. Kedua, Confusion Matrix (peta panas) untuk melihat secara spesifik berapa email aman yang tak sengaja dituduh sebagai spam, atau sebaliknya.
